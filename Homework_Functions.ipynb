```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Узнать владельца документа по его номеру
def doc_owner(n):
    for i in documents:
        if n == i['number']:
            print(f"Владелец документа: {i['name']}")
            return None
    print('Документ не найден в базе')
    return None

#doc_owner(user_input)

# Узнать, на какой полке хранится документ
def shelf_number(n):
    for shelf, number in directories.items():
        if n in number:
            print(f"Документ хранится на полке: {shelf}")
            return None
    print('Документ не найден в базе')
    return None

#shelf_number(user_input)

# Полная информация по всем документам
def complete_info():
    for i in documents:
        for shelf, number in directories.items():
            if i['number'] in number:
                x = shelf
        print(f'№: {i["number"]}, тип: {i["type"]}, владелец: {i["name"]}, полка хранения: {x}')
    return None
            
#complete_info()

# Добавить новую полку
def add_shelf(k):
        if k not in directories.keys():
            directories[k] = ""
            print(f'Полка добавлена. Текущий перечень полок: {", ".join(directories.keys())}')
        else:
            print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(directories.keys())}')
            return None
            
#add_shelf(user_input_shelf)

# Удалить существующую полку из данных (только если она пустая)
def delete_shelf(k):
    if k not in directories.keys():
        print(f'Такой полки не существует. Текущий перечень полок: {", ".join(directories.keys())}')
    elif directories[k] != []:
        print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {", ".join(directories.keys())}')
    elif directories[k] == []:
        del(directories[k])
        print(f'Полка удалена. Текущий перечень полок: {", ".join(directories.keys())}')
        return None

#delete_shelf(user_input_shelf)

# Пользовательский ввод
def main():
    while True:
        user_input_command = input('Введите команду: ')
        if user_input_command == 'p':
            user_input = input('Введите номер документа: ')
            doc_owner(user_input)
        elif user_input_command == 's':
            user_input = input('Введите номер документа: ')
            shelf_number(user_input)
        elif user_input_command == 'l':
            complete_info()
        elif user_input_command == 'ads':
            user_input_shelf = input('Введите номер полки: ')
            add_shelf(user_input_shelf)
        elif user_input_command == 'ds':
            user_input_shelf = input('Введите номер полки: ')
            delete_shelf(user_input_shelf)
        elif user_input_command == 'q':
            print('Конец программы')
            break
        
main()
```
